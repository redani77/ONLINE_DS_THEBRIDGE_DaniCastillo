### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python

print(california_cities_copy.head(50))
```

```python
top_populated_cities = california_cities_copy.nlargest(10, 'population_total')
print(top_populated_cities[['city', 'population_total']])
```

```python
# 1) ciudades más pobladas
top_cities = california_cities_cleaned.nlargest(5, 'population_total')

# (Colores: azul, gris, aceituna, marrón, con trasparencia)
colors = ['blue', 'gray', 'olive', 'brown', (0.5, 0.5, 0.5, 0.5)]  # Transparente al final

# grafico barras
plt.figure(figsize=(8, 6))
bars = plt.bar(top_cities['city'], top_cities['population_total'], color=colors)

# poner los num de población encima de las barras
for bar, population in zip(bars, top_cities['population_total']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f"{population:,}", ha='center', va='bottom', fontsize=10)

# titulos y nombres del grafico
plt.title("Ciudades más pobladas de California")
plt.ylabel("Número de habitantes")
plt.show()
```

```python
top_area_cities = california_cities_copy.nlargest(10, 'area_total_km2')
print(top_area_cities[['city', 'area_total_km2']])
```

```python
# 2) histograma del área total
plt.figure(figsize=(8, 6))
plt.hist(california_cities_cleaned['area_total_km2'], bins=50, color='steelblue', edgecolor='black', alpha=0.7)

# titulos y nombres del grafico
plt.title("Histograma del área total ocupada por las ciudades", fontname="Bradley Hand ITC", fontsize=18)
plt.xlabel("Superficie (km²)")
plt.ylabel("Número de ciudades")
plt.show()
```

```python
water_area_population = california_cities_copy[['city', 'area_water_km2', 'population_total']]
print(water_area_population)
```

```python

# 3) scatter de área de agua de un ciudad con su poblacion
plt.figure(figsize=(8, 6))
plt.scatter(california_cities_cleaned['area_water_km2'], california_cities_cleaned['population_total'],
            color='green', alpha=0.6, edgecolors='black')

# posibles outliers
outliers = california_cities_cleaned[
    (california_cities_cleaned['area_water_km2'] > 80) & (california_cities_cleaned['population_total'] > 3000)
]

# líneas y etiquetar los outliers
for _, row in outliers.iterrows():
    plt.plot([row['area_water_km2'], row['area_water_km2']],
             [0, row['population_total']], color='black', linestyle='--', linewidth=1)
    plt.text(row['area_water_km2'], row['population_total'], "Outliers", fontsize=10)


# titulos y nombres del grafico
plt.title("Población por kilómetro cuadrado de agua")
plt.xlabel("Área de agua en la ciudad (km²)")
plt.ylabel("Población")
plt.grid(alpha=0.3)
plt.show()
```

```python
city_elevation = california_cities_copy[['city', 'elevation_m']]
print(city_elevation)
```

```python
#boxplot de la altura de las ciudades.
plt.figure(figsize=(8, 6))
sns.boxplot(x=california_cities_cleaned['elevation_m'], color='red')

# Personalizar el gráfico
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel("Altitud (m)")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Gráfico de barras
axes[0, 0].bar(top_cities['city'], top_cities['population_total'], color=colors)
for bar, population in zip(axes[0, 0].patches, top_cities['population_total']):
    axes[0, 0].text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                    f"{population:,}", ha='center', va='bottom', fontsize=10)
axes[0, 0].set_title("Ciudades más pobladas de California")
axes[0, 0].set_ylabel("Número de habitantes")

# Histograma
axes[0, 1].hist(california_cities_cleaned['area_total_km2'], bins=50, color='steelblue', edgecolor='black', alpha=0.7)
axes[0, 1].set_title("Histograma del área total ocupada por las ciudades", fontname="Bradley Hand ITC", fontsize=12)
axes[0, 1].set_xlabel("Superficie (km²)")
axes[0, 1].set_ylabel("Número de ciudades")

# Scatterplot
axes[1, 0].scatter(california_cities_cleaned['area_water_km2'], california_cities_cleaned['population_total'],
                   color='green', alpha=0.6, edgecolors='black')
for _, row in outliers.iterrows():
    axes[1, 0].plot([row['area_water_km2'], row['area_water_km2']],
                    [0, row['population_total']], color='black', linestyle='--', linewidth=1)
    axes[1, 0].text(row['area_water_km2'], row['population_total'], "Outliers", fontsize=10)
axes[1, 0].set_title("Población por kilómetro cuadrado de agua")
axes[1, 0].set_xlabel("Área de agua en la ciudad (km²)")
axes[1, 0].set_ylabel("Población")
axes[1, 0].grid(alpha=0.3)
# Boxplot
sns.boxplot(ax=axes[1, 1], x=california_cities_cleaned['elevation_m'], color='red')
axes[1, 1].set_title("Diagrama de caja de la Altitud de las ciudades de California")
axes[1, 1].set_xlabel("Altitud (m)")

plt.tight_layout()
plt.show()
```
